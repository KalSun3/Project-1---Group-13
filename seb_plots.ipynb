```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
file_path = "Resources/ds_salaries.csv"
data = pd.read_csv(file_path)
```

```python
pd.set_option('display.max_rows', None)
styled_data = data.style.set_table_styles([
    {'selector': 'th', 'props': [('font-size', '12pt'), ('text-align', 'center')]},
    {'selector': 'td', 'props': [('font-size', '10pt'), ('text-align', 'center')]}
])

styled_data
```

```python
salary = data['salary_in_usd']
years = data['work_year']
plt.scatter(salary,years)
plt.xlabel("Salary in USD ($)")
plt.ylabel("Years")
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 2)
sns.boxplot(x='work_year', y='salary_in_usd', data=data, palette="Paired", legend=False)
plt.title('Salary in USD to Year comparison')
plt.xlabel('Year of Work')
plt.ylabel('Salary in USD')
#plt.legend(handles, employment_labels.values(), title='Employment Type')
#plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 2)
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=data, palette="Paired", legend=False)
plt.title('Salary in USD to Year comparison')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Salary In USD')
#plt.legend(handles, employment_labels.values(), title='Employment Type')
#plt.tight_layout()
plt.show()
```
